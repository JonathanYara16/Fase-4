## Análisis Multivariado

```python
# Calcular el rango intercuartil (IQR) para poder hallar Outliers
Q1 = data.quantile(0.25)  # 1er cuartil
Q3 = data.quantile(0.75)  # 3er cuartil
IQR = Q3 - Q1  # Fórmula de rango intercuartil

# Identificar outliers (valores fuera del rango [Q1 - 1.5 * IQR, Q3 + 1.5 * IQR])
outliers = (data < (Q1 - 1.0 * IQR)) | (data > (Q3 + 1.0 * IQR))

# Mostrar filas con Outliers 
print(f"Filas originales: {data.shape[0]}")
outliers_rows = data[outliers.any(axis=1)]  # Filtrar las filas con outliers

# Mostrar solo la cantidad de filas con outliers
print(f"Cantidad de filas con outliers: {outliers_rows.shape[0]}")

# Eliminar los outliers
data_no_outliers = data[~outliers.any(axis=1)]

# Mostrar tabla con las filas sin outliers
print(f"\nFilas sin outliers: {data_no_outliers.shape[0]}")
print(data_no_outliers)
```

```python
# Selección de variables relevantes sin datos atípicos dentro de la BD
data_clustering_no_outliers = data_no_outliers[['Gender', 'Age', 'Annual Income (k$)', 'Spending Score (1-100)']]

# Verificación que la selección de columnas se haya hecho correctamente
#print(f"Columnas seleccionadas para clustering: {data_clustering.columns}")
```

```python
# Normalización de data_Clustering sin datos atípicos 
scaler = StandardScaler()
data_normalized_no_outliers = scaler.fit_transform(data_clustering_no_outliers)
#data_normalized = scaler.fit_transform(data_clustering)

# Convertir los datos normalizados a un DataFrame para visualización
data_normalized_no_outliers_df = pd.DataFrame(data_normalized_no_outliers, columns=data_clustering_no_outliers.columns)

# Visualiza las primeras 5 filas de los datos normalizados
print(data_normalized_no_outliers_df.head())
```

```python
points_array = data.to_numpy()
def plot_elbow_method(data_normalized_no_outliers, max_clusters=12):
    inertias = []
    
    for i in range(1, max_clusters+1):
        kmeans = KMeans(n_clusters=i, random_state=42)
        kmeans.fit(data_normalized_no_outliers)
        inertias.append(kmeans.inertia_)
    
    plt.figure(figsize=(7, 5))
    plt.plot(range(1, max_clusters+1), inertias, marker='o', linestyle='-', color='b')
    plt.title('Método del Codo')
    plt.xlabel('Número de Clusters')
    plt.ylabel('Inercia')
    plt.grid(True)
    plt.show()

# Llama a la función para visualizar el codo
plot_elbow_method(data_normalized_no_outliers, max_clusters=12)
```

```python


# Número de clusters a evaluar
num_clusters = 7

kmeans = KMeans(n_clusters=num_clusters, random_state=42)
clusters = kmeans.fit_predict(data_normalized_no_outliers)

# Coeficiente de Silhouette
silhouette_avg = silhouette_score(data_normalized_no_outliers, clusters)

# Índice Calinski-Harabasz
calinski_harabasz_index = calinski_harabasz_score(data_normalized_no_outliers, clusters)

# Crear los gráficos de los dos índices
fig, axs = plt.subplots(1, 2, figsize=(15, 5))

# Gráfico del Coeficiente de Silhouette
axs[0].plot([num_clusters], [silhouette_avg], marker='o', color='r')
axs[0].set_xlabel('Número de Clusters')
axs[0].set_ylabel('Coeficiente de Silhouette')
axs[0].set_title('Coeficiente de Silhouette')

# Gráfico del Índice Calinski-Harabasz
axs[1].plot([num_clusters], [calinski_harabasz_index], marker='o', color='g')
axs[1].set_xlabel('Número de Clusters')
axs[1].set_ylabel('Índice Calinski-Harabasz')
axs[1].set_title('Índice Calinski-Harabasz')

plt.tight_layout()
plt.show()
```

```python
# Definir el rango de números de clusters a probar - linea de trayectoria 
range_n_clusters = range(2, 14)  # Prueba 13 clusters

# Inicializar listas para almacenar los resultados
silhouette_avg = []
calinski_harabasz = []

# Calculamos índices para cada número de clusters
for n_clusters in range_n_clusters:
    # Realizar el clustering con KMeans
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    clusters = kmeans.fit_predict(data_normalized_no_outliers)
    
# Calcular el coeficiente de Silhouette
 silhouette_avg.append(silhouette_score(data_normalized_no_outliers, clusters))
    
# Calcular el índice Calinski-Harabasz
calinski_harabasz.append(calinski_harabasz_score(data_normalized_no_outliers, clusters))

# Crear los gráficos de los dos índices
fig, axs = plt.subplots(1, 2, figsize=(18, 6))

# Gráfico del Coeficiente de Silhouette con línea
axs[0].plot(range_n_clusters, silhouette_avg, marker='o', color='r', linestyle='-', label="Silhouette")
axs[0].set_xlabel('Número de Clusters')
axs[0].set_ylabel('Coeficiente de Silhouette')
axs[0].set_title('Coeficiente de Silhouette')
axs[0].legend()

# Gráfico del Índice Calinski-Harabasz con línea
axs[1].plot(range_n_clusters, calinski_harabasz, marker='o', color='g', linestyle='-', label="Calinski-Harabasz")
axs[1].set_xlabel('Número de Clusters')
axs[1].set_ylabel('Índice Calinski-Harabasz')
axs[1].set_title('Índice Calinski-Harabasz')
axs[1].legend()

plt.tight_layout()
plt.show()
```

```python
# Realizar el clustering jerárquico utilizando el método de Ward - Generado Automaticamente
linked = linkage(data_normalized_no_outliers, method='ward')

# Creación del dendrograma
plt.figure(figsize=(7, 5))
dendrogram(linked)
plt.title('Dendrograma - Clustering Jerárquico')
plt.xlabel('Índice del Cliente')
plt.ylabel('Distancia Euclidiana')
plt.show()
```

```python
# Realizar el clustering jerárquico utilizando el método de Ward - Generado con 7 Clusters
linked = linkage(data_normalized_no_outliers, method='ward')

# Creación del dendrograma
plt.figure(figsize=(7, 5))
dendrogram(linked)
plt.title('Dendrograma - Clustering Jerárquico')
plt.xlabel('Índice del Cliente')
plt.ylabel('Distancia Euclidiana')
plt.show()

num_clusters_jerarquico = 7
clusters_jerarquico = fcluster(linked, num_clusters_jerarquico, criterion='maxclust')

# Agregar los clusters como una nueva columna al DataFrame original
data_no_outliers['Cluster'] = clusters_jerarquico

# Mostrar las primeras filas con los clusters asignados
print(data_no_outliers.head())
```

```python
pca = PCA(n_components=2)
data_pca = pca.fit_transform(data_normalized_no_outliers)
Clusters = data_no_outliers['Cluster']


# Gráfico de dispersión con colores para cada cluster
plt.figure(figsize=(6, 4))
scatter = plt.scatter(data_pca[:, 0], data_pca[:, 1], c=Clusters, cmap='viridis', s=50)
plt.colorbar(scatter, label='Cluster')
plt.title('Visualización de Clusters en 2D (PCA)')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.show()
```

```python
#Realizamos la grafica de dispesión con 7 clusters - quitando dimensiones
from sklearn.manifold import TSNE 
# Aplicar t-SNE para reducir la dimensionalidad a 2 dimensiones
tsne = TSNE(n_components=2, random_state=42)
data_tsne = tsne.fit_transform(data_normalized_no_outliers)

Clusters = data_no_outliers['Cluster']

# Gráfico de dispersión con colores para cada cluster
plt.figure(figsize=(6, 4))
scatter = plt.scatter(data_tsne[:, 0], data_tsne[:, 1], c=Clusters, cmap='viridis', s=50)
plt.colorbar(scatter, label='Cluster')
plt.title('Visualización de Clusters en 2D (t-SNE)')
plt.xlabel('Componente 1 (t-SNE)')
plt.ylabel('Componente 2 (t-SNE)')
plt.show()
```

```python
# Obtener el tamaño de los clusters para k=7 y k=10
kmeans_7 = KMeans(n_clusters=7, random_state=42).fit(data_normalized_no_outliers)
kmeans_8 = KMeans(n_clusters=8, random_state=42).fit(data_normalized_no_outliers)

# Contar el tamaño de cada cluster
sizes_7 = np.bincount(kmeans_7.labels_)
sizes_8 = np.bincount(kmeans_8.labels_)


print(f"Tamaño de los clusters con k=7: {sizes_7}")
print(f"Tamaño de los clusters con k=8: {sizes_8}")


# Visualizar los clusters para k=7 y k=10 en un gráfico 2D usando PCA
pca = PCA(n_components=2)

# PCA para k=7
data_pca_7 = pca.fit_transform(data_normalized_no_outliers)
plt.figure(figsize=(5, 3))
plt.scatter(data_pca_7[:, 0], data_pca_7[:, 1], c=kmeans_7.labels_, cmap='viridis')
plt.title("Clustering con k=7")
plt.xlabel('Componente principal 1')
plt.ylabel('Componente principal 2')
plt.show()

# PCA para k=8
data_pca_10 = pca.fit_transform(data_normalized_no_outliers)
plt.figure(figsize=(5, 3))
plt.scatter(data_pca_10[:, 0], data_pca_10[:, 1], c=kmeans_8.labels_, cmap='viridis')
plt.title("Clustering con k=8")
plt.xlabel('Componente principal 1')
plt.ylabel('Componente principal 2')
plt.show()
```
